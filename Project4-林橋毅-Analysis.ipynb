Accruacy

```python
N = 10000
```

```python
correct_avg = 0
wrong = 0
random_avg = 0
for i in range(N):
    f_correct = open("/Users/linqiaoyi/NCU/CE2034/flowPrj/env01/maxFlow_ans/" + str(i) + ".out")
    f_random_500 = open("/Users/linqiaoyi/NCU/CE2034/flowPrj/env01/random_ans_2500/" + str(i) + ".out")
    correct_ans = f_correct.readline()
    random_ans = f_random_500.readline()
    if(correct_ans != random_ans):
 #       print(correct_ans, random_ans)
        wrong += 1
    correct_time = float(f_correct.readline())
    correct_avg += correct_time
    random_time = float(f_random_500.readline())
    random_avg += random_time
print(f'The accuracy is {(N-wrong)/N * 100} %')
print(correct_avg / N * 1000, random_avg / N * 1000)
```

```python
N = 995
correct = []
random_V = []
random_V2 = []
for i in range(2,201):
    correct_time = 0
    random_V_time = 0
    random_V2_time = 0
    for j in range(5):
        f_correct = open("/Users/linqiaoyi/NCU/CE2034/flowPrj/env02/maxFlow_ans/"+ str(5*(i-2)+j) + ".out")
        f_random_V = open("/Users/linqiaoyi/NCU/CE2034/flowPrj/env02/random_ans/"+ str(5*(i-2)+j) + ".out")
        f_random_V2 = open("/Users/linqiaoyi/NCU/CE2034/flowPrj/env02/random_ans_V2/"+ str(5*(i-2)+j) + ".out")
        correct_ans = f_correct.readline()
        randomV_ans = f_random_V.readline()
        randomV2_ans = f_random_V2.readline()

        correct_time += float(f_correct.readline()) * 1000
        random_V_time += float(f_random_V.readline()) * 1000
        random_V2_time += float(f_random_V2.readline()) * 1000
    
    correct.append(correct_time/5)
    random_V.append(random_V_time/5)
    random_V2.append(random_V2_time/5)
```

```python
plt.rcParams["font.sans-serif"]=["Arial Unicode MS"]
plt.title("固定稀疏程度 50 % 執行時間比較圖")
plt.xlabel("點的數量（個）")
plt.ylabel("平均執行時間(ms)")
plt.plot(range(2,201), correct, label = "Max-Flow 解法")
plt.plot(range(2,201), random_V, label= "隨機做 V 次")
plt.legend()
plt.show()
```

```python
correct = []
random50 = []
random2500 = []
for i in range(5,101):
    correct_time = 0
    random50_time = 0
    random2500_time = 0
    for j in range(5):
        f_correct = open("/Users/linqiaoyi/NCU/CE2034/flowPrj/env03/maxFlow_ans/"+ str(5*(i-5)+j) + ".out")
        f_random_50 = open("/Users/linqiaoyi/NCU/CE2034/flowPrj/env03/random_ans_50/"+ str(5*(i-5)+j) + ".out")
        f_random_2500 = open("/Users/linqiaoyi/NCU/CE2034/flowPrj/env03/random_ans_2500/"+ str(5*(i-5)+j) + ".out")
        
        correct_ans = f_correct.readline()
        random50_ans = f_random_50.readline()
        random2500_ans = f_random_2500.readline()
        
        correct_time += float(f_correct.readline())
        random50_time += float(f_random_50.readline())
        random2500_time += float(f_random_2500.readline())
        
    correct.append(correct_time / 5 * 1000)
    random50.append(random50_time / 5 * 1000)
    random2500.append(random2500_time / 5 * 1000)
```
